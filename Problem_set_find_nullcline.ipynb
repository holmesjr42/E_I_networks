```python
# for fast array manipulation
import numpy as np
# for plotting
import matplotlib.pyplot as plt
# for numerical ODE integration
from scipy.integrate import odeint
# for nonlinear equations
from scipy.optimize import fsolve
from matplotlib.pyplot import cm
#find fixed points with scipy?
import scipy.optimize as optimize

# to display plots in-line
%matplotlib inline

#import sympy for fixed points?
import sympy as sm

# minimum and maximum E and I values we want displayed in the graph-sayshttp://martinosorb.github.io#variables
tau=0.01
w_ee=15
w_ei=-15
w_ie=10
w_ii=-5
e=0


#define tanh np.maximum

def wow(x):
    np.maximum(0,x)
    return np.tanh(x)

En=[]
In=[]

#nullcline value
def fn1(En,In):
    return (-En+wow(w_ee*En+w_ei*In+e))/tau
def gn1(En,In):
    return (-In+wow(w_ie*En+w_ii*In))/tau

Encx=[]
Incx=[]
Ency=[]
Incy= []
def nullcline(r):
    for En in range(-r,r):
        for In in range(-r,r):
            if ((fn1(En,In) == 0)):
                Encx.append(En)
                Incx.append(In)
                
            if ((gn1(En,In)==0)):
                Ency.append(En)
                Incy.append(In)
    nin=2*r
    for ninn in range(nin):
        print([Encx[ninn],Incx[ninn]],':',[Ency[ninn],Incy[ninn]])
        


```

```python
nullcline(100) 
plt.figure(figsize=(10,10))
plt.xlim([-1,1])
plt.ylim([-1,1])
plt.plot(Incx, Encx,'b--')
plt.plot(Incy, Ency,'c.-')
```
